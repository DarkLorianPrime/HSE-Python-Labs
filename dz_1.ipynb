Первое задание: проверка високосности года

```python
year = 2013
```

Если требуется проверка на то, что год должен быть четырехзначным - выполнить блок ниже (по-умолчанию выполнен)

```python
if not (999 < year < 10000):
    print("Принимаются только четырехзначные числа")
    raise ValueError("Число должно быть от 1000 до 9999")
```

```python
if (year % 4 == 0) and (year % 100 != 0) or (year % 400 == 0):
    print("Високосный год")
else:
    print("Обычный год")
```

Второе задание: проверка билета на счастливый номер

```python
ticket = "123322" # Или это могло бы быть инпутом
if len(ticket) != 6 or not ticket.isdigit():
    print("Номер билета должен быть шестизначной строкой состоящей из цифр")
    raise ValueError("ticket должен быть длиной в 6 символов и состоять из цифр")
```

```python
first_half = sum(map(int, ticket[:3]))
second_half = sum(map(int, ticket[3:]))
```

```python
if first_half == second_half:
    print("Счастливый билет")
else:
    print("Несчастливый билет")
```

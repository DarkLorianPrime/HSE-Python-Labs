Задание 1:
Пользователь по команде p может узнать владельца документа по его номеру.

Задание 2:
Секретарь по команде s может по номеру документа узнать, на какой полке документ хранится.


```python
import sys

documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
def get_owner_by_document_number(doc_number):
    """Найти владельца документа по номеру"""
    for doc in documents:
        if doc['number'] == doc_number:
            return doc['name']
    return None


def get_shelf_by_document_number(doc_number):
    """Найти полку, на которой хранится документ"""
    for shelf, docs in directories.items():
        if doc_number in docs:
            return shelf
    return None
```

```python
print("Программа для работы с документами")
print("Доступные команды:")
print("p - найти владельца документа")
print("s - найти полку документа")
print("q - выход")
```

```python
command = input("\nВведите команду: ").strip().lower()

if command == 'q':
    print("Выход из программы.")
    sys.exit(0)

elif command == 'p':
    doc_number = input("Введите номер документа: ").strip()
    owner = get_owner_by_document_number(doc_number)

    if owner:
        print(f"Владелец документа: {owner}")
    else:
        print("Владелец документа: владелец не найден")
elif command == 's':
    doc_number = input("Введите номер документа: ").strip()
    shelf = get_shelf_by_document_number(doc_number)

    if shelf:
        print(f"Документ хранится на полке: {shelf}")
    else:
        print("Документ не найден на полках")

else:
    print("Неизвестная команда. Попробуйте снова.")
```

```python
import datetime
import sys
```

```python
allowed_formats_dict = {
    "the moscow times": "%A, %B %d, %Y",
    "the guardian": "%A, %d.%m.%y",
    "daily news": "%A, %d %B %Y"
}
```

Блок кода снизу представляет собой инпут для приведения текущей даты и времени к формату журналов.

```python
format = input(f"Введите один из {len(allowed_formats_dict)} форматов или q для выхода. Доступные форматы: {', '.join(allowed_formats_dict.keys())}")
if format == "q":
    sys.exit(0)
elif format not in allowed_formats_dict.keys():
    print("Формат не поддерживается.")
    sys.exit(1)
```

Снизу представлен код вывода текущего времени в заданном формате

```python
print(format, "—", datetime.datetime.now().strftime(allowed_formats_dict[format]))
```

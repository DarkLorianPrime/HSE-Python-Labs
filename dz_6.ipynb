```python
import csv
```

```python
purchases = {}
```

```python
with open('visit_log.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader)  # пропускаем заголовок

    for row in reader:
        if len(row) >= 3:
            user_id, source, category = row[0], row[1], row[2]

            # Если есть категория (покупка), сохраняем её
            if category and category != '':
                purchases[user_id] = category
```

```python
with open('visit_log.csv', 'r', encoding='utf-8') as input_file, open('funnel.csv', 'w', newline='',
                                                                      encoding='utf-8') as output_file:
    reader = csv.reader(input_file)
    writer = csv.writer(output_file)

    # Записываем заголовок
    headers = next(reader)
    writer.writerow(['user_id', 'source', 'category'])

    # Обрабатываем каждую строку
    for row in reader:
        if len(row) >= 2:
            user_id, source = row[0], row[1]

            # Если у пользователя была покупка, записываем строку
            if user_id in purchases:
                writer.writerow([user_id, source, purchases[user_id]])
```
